Use clustering methods to work out if there are distinct types of asteroid

```python
# standardize data
from sklearn import preprocessing
standardized_data = preprocessing.scale(cleaning_data.loc[:,'Alamosite':'Kostovite'])
standardized_data_df = pd.DataFrame(standardized_data)
```

```python
from sklearn import metrics
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
def k_silhouette(X,clusters):
    K = range(2, clusters+1)
    S = []
    for k in K:
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(X)
        labels = kmeans.labels_
        S.append(metrics.silhouette_score(X,labels,metric='euclidean'))
    plt.plot(K,S,'b*-')
    plt.xlabel('k')
    plt.ylabel('silhouette_score')
    
    plt.show();
k_silhouette(standardized_data_df,20)
```

```python
sse = []
for k in range(1,20):
        kmeans = KMeans(n_clusters=k,)
        kmeans.fit(standardized_data_df)
        sse.append(kmeans.inertia_)
x = range(1,20)
plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(x,sse,'o-')
plt.show()
```

```python
data = pd.read_csv('China Aug 2022 Asteroid 2 v4.csv')
```

```python
data
```

```python
df = pd.DataFrame(data).apply(pd.to_numeric, errors='coerce')
df
```

```python
for column_name in df:
    df[column_name].fillna(df[column_name].mean(),inplace=True)
df
```

```python
print(df.isnull().sum())
```

```python
df.describe()
```

```python
df = df.astype('int32' , errors='ignore')
```

Category is the label, use supervised learning methods

Decision tree:

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
```

```python
x = df.loc[:,'Chalcophyllite':'Diadochite']
x
```

```python
x.info()
```

```python
y = df.loc[:,'Category':'Category']
y.info
```

```python
model = DecisionTreeClassifier(random_state=0, max_depth=3, ccp_alpha=0.001) # Tune
```

```python
model.fit(x,y)
```

```python
dot_data = export_graphviz(model, out_file = 'mechanically_stable.dot',
               feature_names = x.columns,
               class_names = ['Mechanically stable','NOT meachanically stable'],
               rounded = True, proportion = False,
               precision = 2, filled = True)
```

Random Forest classifer

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
# Best result after Tune
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
rfc.score(x_test,y_test)
```

##### After improving, Random forest classifier performance is around 0.95.

```python
# from sklearn.metrics import plot_confusion_matrix
# plot_confusion_matrix(rfc,x_test,y_test,display_labels=['Mechanically stable','NOT meachanically stable'],cmap=plt.cm.Blues,)
# plt.show()
```

```python
# Best result after Tune (max_depth, ccp_alpha)
dtc = DecisionTreeClassifier(random_state=0)
dtc.fit(x_train,y_train)
dtc.score(x_test,y_test)
```

##### After improving, Decision tree performance is around 0.85.
